```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('/content/Salary_Data.csv')
```

```python
data.head()
```

```python
#according to experience we have the salary.
```

```python
#This data doesn't require any type of preprocessing. Because no null values.
#we have to divide data into features and target i.e x nd y.
# so here the x is YearsExperience and y is Salary
```

```python
x=data['YearsExperience']
y=data['Salary']
```

```python
x
```

```python
type(x)
```

```python
#x is a series that is one dimensional data.
#x should be array like data. For good model prediction.
#so x should be a 2 dimenional array or a dataframe not a series or 1-d array.
#y can be 1d or 2d both.
# so to make x as a 2d array

x= data[['YearsExperience']]
```

```python
x.shape
```

```python
type(x)
```

```python
print(x.shape)
print(y.shape)
```

```python
# now x is my feature and y is my target.
```

```python
#Data is now clean and fine
#Now the first step in Ml is split the data in Train and test.
#so for that i will use a library from scikit learn to tain and test data splitting.

```

```python
from sklearn.model_selection import train_test_split
#will take 4 variables.
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.8,random_state=9014)
#random_state is given to have same random state just like the seed.
#seed is given to fix the randomness.
#range of seed is 0-2^31.
 # I have to split x and y either you can pass traing size or test size other one will be taken automatically
 #xtrain and ytrain have 80% of the data and xtest and ytest have 20% of data.
```

```python
#80% of 30 is 24 hence 6 samples for testing.
```

```python
xtest
```

```python
#now we split data randomly
```

```python
xtrain
```

```python
#after splitting the nest step is picking up the model .That is the Model Selection.
#Model Selection is Linear Regression as two only.
```

```python
#Build the model
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
## Fit the model to train it.
model.fit(xtrain,ytrain)
```

```python
#everything is done now.
#Now if I want to test this to my testing data.
# Testing or Prediction.

ypred=model.predict(xtest)  # we will pass the xtest data to predict y.
ypred
```

```python
ytest.values
```

```python
#There is very less diff between actual and predicted value.
```

```python
#Now the next step is to Calculate R2 score.
#Accuracy?
```

```python
from sklearn.metrics import r2_score
r2=r2_score(ytest,ypred)
r2
```

```python
#r2 is 0.988 which is 98.8% accuracy which is good.
#so we will try to increase it more but before that lets draw the line of regression.
```

```python
#one for training samples and other for testing samples.

#Plot Line of Regression for the Training sample.
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xtrain,ytrain,color='red',s=100,label="Actual Data")
plt.scatter(xtrain,model.predict(xtrain),color='green',s=100,label="Predicted Data")
#x was same only y was changed during fining line of regresion
plt.plot(xtrain,model.predict(xtrain),color='orange',label="Line of Regression")
plt.legend()
plt.show()
```

```python
#same thing we will do for the testing sample.
plt.figure(figsize=(8,6))
plt.scatter(xtest,ytest,color='red',s=100,label="Actual Data")
plt.scatter(xtest,model.predict(xtest),color='green',s=100,label="Predicted Data")
#x was same only y was changed during fining line of regresion
plt.plot(xtest,model.predict(xtest),color='orange',label="Line of Regression")
plt.legend()
plt.show()
```

```python
m=model.coef_
c=model.intercept_
print(m,c)
```
