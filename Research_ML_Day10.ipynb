```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/income.csv')
```

```python
data.head()
```

```python
#Name can't be used for cluster aas string
#so we will use only age and cluster
```

```python
#lets plot graph and checkby looking at the graph the num of cluters.
```

```python
plt.scatter(data['Age'],data['Income($)'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

```python
x=data[['Age','Income($)']]
x
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
ypred= km.fit_predict(x)
ypred
```

```python
x['Cluster']=ypred
x
```

```python
#to get centroid of all clusters
centroids=km.cluster_centers_
#x and y value for each centroid
```

```python
#to plot i have to seperate datapts
df1 = x[x['Cluster']==0]
df1
```

```python
df2 = x[x['Cluster']==1]
df3 = x[x['Cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'],color='green',label="Cluster1", s=100)
plt.scatter(df2['Age'],df2['Income($)'],color='red',label="Cluster1", s=100)
plt.scatter(df3['Age'],df3['Income($)'],color='blue',label="Cluster1", s=100)
plt.scatter(centroids[:,0],centroids[:,1],s=200,marker="*",label="Centroid",color="purple")
#i want all rows and cols
plt.show()
```

```python
#it is errored we are not getting proper clustering
#There is so much diff in values
#scaling should be applied here.
#Apply Scaling
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data['Age']=scaler.fit_transform(data[['Age']])
data['Income($)']=scaler.fit_transform(data[['Income($)']])
```

```python
plt.scatter(data['Age'],data['Income($)'])
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
x=data[['Age','Income($)']]
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
ypred= km.fit_predict(x)
x['Cluster']=ypred
centroids=km.cluster_centers_
df1 = x[x['Cluster']==0]
df2 = x[x['Cluster']==1]
df3 = x[x['Cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'],color='green',label="Cluster1", s=100)
plt.scatter(df2['Age'],df2['Income($)'],color='red',label="Cluster1", s=100)
plt.scatter(df3['Age'],df3['Income($)'],color='blue',label="Cluster1", s=100)
plt.scatter(centroids[:,0],centroids[:,1],s=200,marker="*",label="Centroid",color="purple")
#i want all rows and cols
plt.show()
```

```python
#we get centroid properly and clusters properly
```

```python
#This is how we apply k means clustering
```

```python
#Now use elbow method find value of K
#did we get same k value or not
```

```python
sse=[]
for k in range(1,20):
  km=KMeans(n_clusters=k)
  km.fit(data[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.xlabel("K Value")
plt.ylabel("SSE")
plt.xticks(range(1,20))
plt.plot(range(1,20),sse)
plt.show()
```
