Feauture Encoding:-
First of all we have two type of data.
Categorical data((having Category like Gender col, team and senior management col in prev data))

Ordinal data(which is having some rank or order just like education, Employees(manager, Senior director and so on.)


Another is the Nominal Data.It ha snor rank.(Gender is Nominal data)

Team is Ordinal data.
Gender is not ordinal data.
senior Mangement is ordinal data.

Based on this data we have encoding.
For the Nominal Encoding.
The most used type is One Hot Encoding. It is a type of Encoding we use for Nominal Data.

For the Ordinal data the encoding we use is label encoding.

ONE_HOT_ENCODING:- Here we have a concept of dummy variables. One hot encoding is applied for Nominal Encoding.

Suppose countries USA/China/India.(Encoding means converting Categorical String value to numerical value , we require this because we can't apply functions to String data. One Hot Encoding creates Dummy Variables.)

here we have 3 classes hence 3 cols are created.

One hot Encoding can be done by two ways.
Pandas(get_dummies).

Scikitlearn(one hot encoding).


LABELED ENCODING:-
Education(Btech, Mtech and Phd).
Based on the rank you assign the label.
The degree phd will have the most priority, Rank 2 will be mtech and rank 3 will be btech. Based on rank you assign the label would be the class having highest rank will has the highest label. PHd label 2, Mtech label 1, Btech label=0.

FEATURE SCALING.

means converting the value to some specified range.
Suppose you have data age, Salary and Experience.
age=(25,28,30)
,salary(15000,25000,40000),
experience(1,2,0.5)
age range is 25-30 and Salary range is (15000-40000) and for exp is (0.5 to 2).

Model gives highest priority to salary and then age and at last exp and it is not good.

Hence we scale the data or normalize the data. It is Normalization.

We have Standard Scaler and Min Max Scaler(to Neural Network Part and IMages(range0-255)).

Standard Scalar(other data 99% cases.) After Scaling all feautures mean near to 0, standard deviation near to 1.


Formula for Standard Scaler is âŽ

z=(x-x of mean)/standard deviation.

In Min Max Scaler we convert data to any range you want. (0-1, -1to 1 and so on).

Formula of Min Max Scaler:-
x(scale)=(x-x of min)/(x of max-x of min)


ROBUST SCALAR (for feautures having Outliers).
x(scaled)=(x-x of median)/IQR

IQR is Interquartile Range

OUTLIERS:-

Irrelevant Data. or Boundary Values. or the data which is distant from all other other observations(which is totally different).

Lets say we have age data(12,16,14,12,15,18,20,22,18,20,17,90,22,23,76)
So here the outlier is 90 and 76 they are very distant from normal values. They may give bad accuracy and bad prediction. They need to be removed if they are less ore you can modify them.

Ways to find out Outliers:-
1) You plot a Scatter Plot Graph.
2)Using Box Plot
3)Using z score formula.
4)Using IQR(Inter quantile Range).
